```python
model = 'xgb'

Qmax7Fname = f'{model}_cv10_Qmax7_seasonal4_multi_MSWX_meteo_raw_result.csv'
Qmin7Fanme = f'{model}_cv10_Qmin7_seasonal4_multi_MSWX_meteo_raw_result.csv'

df_Qmax7 = pd.read_csv(f'../results/{Qmax7Fname}')
df_Qmin7 = pd.read_csv(f'../results/{Qmin7Fanme}')

if 'climate_label' not in df_Qmax7.columns:
    df_Qmax7['climate_label'] = df_Qmax7.climate.map({1:'tropical',2:'dry',3:'temperate',4:'cold',5:'polar'})
    df_Qmin7['climate_label'] = df_Qmin7.climate.map({1:'tropical',2:'dry',3:'temperate',4:'cold',5:'polar'})
    
palette = {'tropical':'#F8D347','dry':'#C7B18A','temperate':"#65C2A5",'cold':"#a692b0",'polar':"#B3B3B3"}

outName = re.sub('Qmax7_','',Qmax7Fname).split('.')[0]
```

### plot PDF of the target variable

```python
plt.rcParams['axes.labelsize'] = 10
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
fig, axes = plt.subplots(2, 3, figsize=(8,4))
for i,name in enumerate(['Qmax7','Qmin7']):
    if name == 'Qmax7':
        df = df_Qmax7.copy()
    else:
        df = df_Qmin7.copy()
    df['Qmmday'] = df['Q'] / df['gritDarea'] * 86.4
    if (df.Q==0).any():
        df['Qmmday'] = df['Qmmday'] + 0.1
    df['Qmmday (log)'] = np.log(df['Qmmday'])
    sns.kdeplot(df, x = 'Q', ax = axes[i,0], clip = [0, 1000])
    sns.kdeplot(df, x = 'Qmmday', ax = axes[i,1], clip = [0, 10])
    sns.kdeplot(df, x = 'Qmmday (log)', ax = axes[i,2])
    axes[i,0].set_title(f'{name} (m3/s)', fontsize = 10)
    axes[i,1].set_title(f'{name} (mm/day)', fontsize = 10)
    axes[i,2].set_title(f'{name} log-transformed (mm/day)', fontsize = 10)
fig.tight_layout()
fig.savefig('../picture/hist_Qmin7_Qmax7.png', dpi = 600)
```

### plot a map of the length of records

### plot the sensitivity of model performance to length of records

```python
import seaborn as sns
df = df_Qmin7.copy()
df_sta = df.groupby(['ohdb_id','ohdb_latitude', 'ohdb_longitude','climate_label','gritDarea','country']).apply(
        lambda x: pd.Series(
            he.kge(x.pred.values, x.Q.values).squeeze().tolist()+[
                x.shape[0],np.mean(x.pred.values-x.Q.values)/np.mean(x.Q.values)*100
            ], index = ['KGE','r','alpha','beta','Number of seasonal records','bias'])
    ).reset_index()
fig, axes = plt.subplots(2, 2, figsize = (12, 5))
df_sta['tmp'] = pd.cut(df_sta['Number of seasonal records'], bins = np.arange(70, 170, 10).tolist())
df_sta1 = df_sta.groupby('tmp')[['KGE','alpha','beta']].mean().reset_index()
df_sta['darea'] = pd.qcut(df_sta['gritDarea'], 10)
df_sta2 = df_sta.groupby('darea')['bias'].mean().reset_index()
for i,name in enumerate(['KGE','alpha','beta','bias']):
    ax = axes.ravel()[i]
    if name != 'bias':
        sns.barplot(data = df_sta1, x = 'tmp', y = name, ax = ax)
    else:
        sns.barplot(data = df_sta2, x = 'darea', y = name, ax = ax)
        ax.set_xscale('log')
        ax.set_xlabel('Catchment area (km2)')
fig.savefig(f'../picture/Qmin7_performance_sensitivity_{outName}.png', dpi = 600)
```

### maps of relative bias

```python
fig, axes = plt.subplots(2, 1, figsize = (8, 8.5), subplot_kw={'projection':ccrs.EqualEarth()})
for i,name in enumerate(['Qmin7','Qmax7']):
    if i == 0:
        df = df_Qmin7.copy()
        ax = axes[0]
    else:
        df = df_Qmax7.copy()
        ax = axes[1]
    df_sta = df.groupby(['ohdb_id','ohdb_latitude', 'ohdb_longitude','climate_label']).apply(
        lambda x: pd.Series(
            [np.mean((x.pred.values-x.Q.values)) / np.mean(x.Q.values) * 100], index = ['nRMSE'])
    ).reset_index()
    print(df_sta.loc[df_sta.nRMSE.abs()>100,:].shape[0]/df_sta.shape[0])
    title = f'10-fold cross-validation for {name}'
    label = 'Relative bias (%)'
    if i == 0:
        df_sta['nRMSE1'] = pd.cut(df_sta['nRMSE'], np.arange(-200,240,40), 
                                  labels=["<-160", "[-160,-120]","[-120,-80]","[-80,-40]","[-40,0]",
                                         "[0,40]","[40,80]","[80,120]","[120,160]",">160"])
        sizes = {'[0,40]':1,'[40,80]':2,'[80,120]':4,'[120,160]':16,'>160':64,
                            '[-40,0]':1,'[-80,-40]':2,'[-120,-80]':4,'[-160,-120]':16,'<-160':64}
    else:
        df_sta['nRMSE1'] = pd.cut(df_sta['nRMSE'], np.arange(-100,120,20), 
                                  labels=["<-80", "[-80,-60]","[-60,-40]","[-40,-20]","[-20,0]",
                                         "[0,20]","[20,40]","[40,60]","[60,80]",">80"])
        sizes = {'[0,20]':1,'[20,40]':2,'[40,60]':4,'[60,80]':16,'>80':64,
                            '[-20,0]':1,'[-40,-20]':2,'[-60,-40]':4,'[-80,-60]':16,'<-80':64}
    sns.scatterplot(df_sta, x = 'ohdb_longitude', y = 'ohdb_latitude', hue = 'nRMSE1', size = 'nRMSE1', palette = 'RdBu_r',
                    marker = "$\circ$", ec = "face", alpha = .5, 
                    sizes = sizes, ax = ax, transform = ccrs.PlateCarree())
    sns.move_legend(ax, 'lower left', bbox_to_anchor = (0, 0), title = 'Relative bias (%)', title_fontsize = 9)
    
    ax.set_global()
    ax.set_ylim([-6525154.6651, 8625154.6651]) 
    ax.set_xlim([-12662826, 15924484]) 
    ax.spines['geo'].set_linewidth(0)
    ax.coastlines(linewidth = .2, color = '#707070')
    ax.add_feature(cf.BORDERS, linewidth = .2, color = '#707070')
    ax.text(0, 1, ['a','b'][i], transform = ax.transAxes, ha = 'center', va = 'top', fontsize = 10, weight = 'bold')
    ax.set_title(title, fontsize = 9)
fig.savefig(f'../picture/map_relative_bias_{outName}.png', dpi = 600)
```
