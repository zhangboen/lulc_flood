### add quality check for streamflow observations

```python
def cleanQ(df):
    # eliminate invalid records
    df1 = df.loc[df.Q.apply(lambda x: not isinstance(x, str)),:]
    df2 = df.loc[df.Q.apply(lambda x: isinstance(x, str)),:]
    try:
        df2 = df2.loc[df2.Q.str.match('\d+'),:]
    except:
        pass
    df = pd.concat([df1, df2])
    df['Q'] = df.Q.astype(np.float32)
    return df

def del_unreliableQ(df):
    '''observations less than 0 were flagged as
        suspected, and (b) observations with more than ten consecutive
        equal values greater than 0 were flagged as suspected'''
    df['date'] = pd.to_datetime(df['date'])
    df = df.sort_values('date').set_index('date')
    index = pd.date_range(df.index[0], df.index[-1], freq = 'D')
    df = df.reindex(index).fillna(0)
    df1 = df.diff()
    df1 = df1.where(df1==0, 1).diff()
    start = np.where(df1.values==-1)[0]
    end = np.where(df1.values==1)[0]
    if len(start) == 0 or len(end) == 0:
        # must no less than zero
        df = df.loc[df.Q>=0,:]
        return (df)
    if start[0] > end[0]:
        start = np.array([0]+start.tolist())
    if start[-1] > end[-1]:
        end = np.array(end.tolist()+[df1.shape[0]+10])
    duration = end - start
    start = start[duration>=10]
    end = end[duration>=10]
    del_idx = np.array([item for a,b in zip(start,end) for item in np.arange(a+1,b+2).tolist()])
    del_idx = del_idx[del_idx<df.shape[0]]
    if len(del_idx) > 0:
        df.drop(df.index[del_idx], inplace = True)
    # must no less than zero
    df = df.loc[df.Q>=0,:]
    return (df)

def main(ohdb_id):
    print(ohdb_id)
    df = pd.read_csv(os.environ['DATA']+f'/data/OHDB/OHDB_v0.2.3/OHDB_data/discharge/daily/{ohdb_id}.csv')
    # read
    df = cleanQ(df)
    # quality check
    df = del_unreliableQ(df)
    # count observations for each year
    tmp = df.resample('Y')['Q'].agg(countDay='count', Qmax='max')
    # at least 300 valid samples for each year
    tmp = tmp.loc[tmp.countDay>=300,:]
    # limit to 1981-2014
    tmp = tmp.loc[(tmp.index.year>=1981)&(tmp.index.year<=2014),:].reset_index()
    tmp['ohdb_id'] = ohdb_id
    if tmp.shape[0] >= 10:
        return (tmp)
    else:
        return 

# if __name__ == '__main__':
#     ohdb_ids = gdf.ohdb_id.values
#     pool = mp.Pool(16)
#     df = pool.map(main, ohdb_ids)
#     df = pd.concat(df)
#     df.to_csv('../results/Qmax.csv', index = False)
# tmp = df_train0.loc[~df_train0.ohdb_id.isin(df_train),:]
# ohdb_ids = tmp.ohdb_id.values
# for ohdb_id in ohdb_ids[:10]:
#     a = main(ohdb_id)
#     print(a)
```

### calculate new hydrologic signatures: lag time and no reponse ratio

```python
# Function to identify and delete the smaller sample within 5 days
def delete_smaller_sample(df):
    while True:
        intervals = (df.index[1:] - df.index[:-1]).days
        if (intervals > 5).all():
            break
        indices1 = df.index[:-1][intervals<=5]
        indices2 = df.index[1:][intervals<=5]
        for int1,int2 in zip(indices1,indices2):
            if int1 not in df.index:
                continue
            if df.loc[int1] < df.loc[int2]:
                df.drop(int1, inplace=True)
            else:
                df.drop(int2, inplace=True)
    return df
def calc_lagT(pr_dis, p = 95):
    tmp = pr_dis.unstack(level=-1)
    # use scipy to find local discharge peaks
    R95p = tmp.loc[tmp.pr>=0.1,'pr'].quantile(p/100)
    peaks_pr, _ = find_peaks(tmp.pr.values, height=R95p) # find R95p rainfall peaks
    peaks_dis, _ = find_peaks(tmp.dis.values, height=0)
    # remove peaks within five days
    df_peaks_pr = tmp.iloc[peaks_pr,:].pr
    df_peaks_pr = delete_smaller_sample(df_peaks_pr)
    df_peaks_dis = tmp.iloc[peaks_dis,:].dis
    df_peaks_dis = delete_smaller_sample(df_peaks_dis)
    # calculate lag time bewteen rainfall peaks and discharge peaks
    lagT = []
    noResponse = 0
    for index_pr in df_peaks_pr.index:
        index_dis = df_peaks_dis.loc[df_peaks_dis.index>=index_pr]
        if index_dis.shape[0] == 0:
            noResponse += 1
            continue
        index_dis = index_dis.iloc[[0]].index[0]
        if df_peaks_pr.loc[(df_peaks_pr.index>index_pr)&(df_peaks_pr.index<=index_dis)].shape[0] > 0:
            noResponse += 1
            continue
        lagT.append((index_dis - index_pr).days)
    return (pd.Series([np.mean(lagT), noResponse/df_peaks_pr.shape[0]], 
                      index = ['lagT','noResRatio',]))

from parallel_pandas import ParallelPandas
ParallelPandas.initialize(n_cpu=24, split_factor=12)
# calculate lag time between peak rainfall and peak discharge 
df_combine = pd.concat([df_pr.assign(name='pr'), df_qmm.assign(name='dis')])
df_combine.index.name = 'time'
df_combine = df_combine.reset_index().set_index(['time','name'])
df_lagT = df_combine.p_agg(calc_lagT).T.reset_index()
df_lagT.to_csv('../results/lagT.csv', index = False)
```

### calculate hydrologic signatures

```python
def Eckhardt(Q, alpha=.98, BFI=.80, re=1):
    """
    Recursive digital filter for baseflow separation. Based on Eckhardt, 2004.\n
    Q : array of discharge measurements\n
    alpha : filter parameter\n
    BFI : BFI_max (maximum baseflow index)\n
    re : number of times to run filter
    """
    Q = np.array(Q)
    f = np.zeros(len(Q))
    f[0] = Q[0]
    for t in np.arange(1,len(Q)):
        # algorithm
        f[t] = ((1 - BFI) * alpha * f[t-1] + (1 - alpha) * BFI * Q[t]) / (1 - alpha * BFI)
        if f[t] > Q[t]:
            f[t] = Q[t]
    # calls method again if multiple passes are specified
    return np.nansum(f)/np.nansum(Q)

# read MSWEP rainfall
df_pr = pd.concat([pd.read_parquet(a) for a in glob.glob('../data/MSWEPv280/pr_MSWEP*parquet')])
df_pr['time'] = pd.to_datetime(df_pr['time'])
df_pr = df_pr.sort_values('time')
df_pr = df_pr.set_index('time')
# read GloFAS discharge
import multiprocessing as mp
pool = mp.Pool(8)
df_dis = pd.concat(pool.map(pd.read_parquet, glob.glob('../data/GloFASv3/dis*parquet')))
df_dis['time'] = pd.to_datetime(df_dis['time'])
df_dis = df_dis.sort_values('time')
df_dis = df_dis.set_index('time')
# split stations into North and South 
gdf_sta = read_dataframe('../results/OHDB_station_subset.gpkg', read_geometry = False)
# transform streamflow to specific discharge
df_qmm = pd.concat([df_dis, gdf_sta[['ohdb_id','ohdb_catchment_area']].set_index('ohdb_id').T])
df_qmm.iloc[:-1,:] = df_qmm.iloc[:-1,:].div(df_qmm.iloc[-1,:]) 
df_qmm = df_qmm.iloc[:-1,:]
# calculate hydrologic signatures: Q10_50
Q = df_qmm.quantile([.05, .1, .5, .95])
Q5 = Q.loc[0.05]; Q5.name = 'Q5'
Q95 = Q.loc[0.95]; Q95.name = 'Q95'
Q10_50 = Q.loc[0.1] / Q.loc[0.5]; Q10_50.name = 'Q10_50'
# calculate hydrologic signatures: cv
CV = df_qmm.std() / df_qmm.mean(); CV.name = 'cv'
# Baseflow index
baseflow_index = df_qmm.p_agg(Eckhardt)
baseflow_index.name = 'BFI'
# calculate lag time between peak rainfall and peak discharge 
df_combine = pd.concat([df_pr, df_qmm])
print(df_combine.shape)
lagT = df
# calculate hydrologic signatures: 'FI', 'stream_elas'
FI = []
stream_elas = []
for i,m0 in enumerate([[10,9],[7,6]]):
    hy1 = pd.to_datetime('1991-%02d-01'%m0[0])
    hy2 = pd.to_datetime('2000-%02d-30'%m0[1])
    if i == 0:
        da_qmm0 = df_qmm.loc[:,df_qmm.columns.isin(gdf_sta.loc[gdf_sta.ohdb_latitude>=0,'ohdb_id'])]
        da_p0 = df_pr.loc[:,df_pr.columns.isin(gdf_sta.loc[gdf_sta.ohdb_latitude>=0,'ohdb_id'])]
    else:
        da_qmm0 = df_qmm.loc[:,df_qmm.columns.isin(gdf_sta.loc[gdf_sta.ohdb_latitude<0,'ohdb_id'])]
        da_p0 = df_pr.loc[:,df_pr.columns.isin(gdf_sta.loc[gdf_sta.ohdb_latitude<0,'ohdb_id'])]

    # transform time to hydrologic-cycle time
    da_qmm0 = da_qmm0.loc[(da_qmm0.index>=hy1)&(da_qmm0.index<=hy2),:]
    da_qmm0.index = pd.date_range('1991-1-1', periods = da_qmm0.shape[0], freq = 'D')

    da_p0 = da_p0.loc[(da_p0.index>=hy1)&(da_p0.index<=hy2),:]
    da_p0.index = pd.date_range('1991-1-1', periods = da_p0.shape[0], freq = 'D')

    # calculate mean annual rainfall and flashiness index
    p_mean0 = da_p0.groupby(da_p0.index.year).sum().mean()
    FI0 = da_qmm0.groupby(da_qmm0.index.year).p_agg(lambda x: (np.abs(x.diff()).sum()/x.sum())).mean()
    FI.append(FI0)
    # calculate stream_elas
    mq_tot = da_qmm0.mean()
    mp_tot = da_p0.mean()
    mq = da_qmm0.groupby(da_qmm0.index.year).mean()
    mp = da_p0.groupby(da_p0.index.year).mean()
    dp = mp - mp_tot
    dq = mq - mq_tot
    stream_elas0 = ((dq/mq_tot)/(dp/mp_tot)).median()
    stream_elas.append(stream_elas0)
# merge
FI = pd.concat(FI)
stream_elas = pd.concat(stream_elas)
df_hs = pd.concat([FI, stream_elas, baseflow_index, Q10_50, CV], axis = 1)
df_hs.columns = ['FI','stream_elas','BFI','Q10_50','cv','lagT']
df_hs = df_hs.dropna()
df_hs.index.name = 'ohdb_id'
df_hs = df_hs.reset_index()
df_hs.to_csv('../results/hydrological_signatures_training.csv', index = False)
```

### read and merge predictors

```python
def readClimate(model, ens, scenario, year1, year2):
    df_climate = []
    for var in ['tasmin','tasmax','psl','sfcWind','snw','hurs','mrsos','rsds','pr']:
        if var == 'snw':
            fname1 = glob.glob(f'../data/CMIP6_{scenario}/{var}_*{model}_*{ens}*201412_yearmean_train.parquet')[0]
            df0 = pd.read_parquet(fname1).assign(name=var)
            fname1 = glob.glob(f'../data/CMIP6_{scenario}/{var}_*{model}_*{ens}*201412_snowMelting_yearmax_train.parquet')[0]
            df1 = pd.read_parquet(fname1).assign(name='snwMmax')
            fname1 = glob.glob(f'../data/CMIP6_{scenario}/{var}_*{model}_*{ens}*201412_snowMelting_yearmean_train.parquet')[0]
            df2 = pd.read_parquet(fname1).assign(name='snwMmean')
            df0 = pd.concat([df0,df1,df2], axis = 0)
        elif var == 'pr':
            fnames = glob.glob(f'../data/CMIP6_{scenario}/{var}_*{model}_*{ens}*train.parquet')
            fnames = [a for a in fnames if a.split('_')[-2] in ['annsum','pvol','rx1day','rx5day','rx1mon']]
            df0 = pd.concat([pd.read_parquet(fname).assign(name=fname.split('_')[-2]) for fname in fnames], axis = 0)
        else:
            fname1 = glob.glob(f'../data/CMIP6_{scenario}/{var}_*{model}_*{ens}*yearmean_train.parquet')[0]
            df0 = pd.read_parquet(fname1).assign(name=var)
        df_climate.append(df0)
    df_climate = pd.concat(df_climate)
    df_climate = df_climate.set_index('time')
    df_climate.index = pd.to_datetime(df_climate.index)
    df_climate = df_climate.loc[(df_climate.index.year>=year1)&(df_climate.index.year<=year2),:]
    # calculate mean and std
    df_hist_mean = df_climate.groupby('name').mean().T
    df_hist_std = df_climate.groupby('name').std().T
    df_hist_std.columns = [a+'_std' for a in df_hist_std.columns]
    df_hist = pd.concat([df_hist_mean, df_hist_std], axis = 1)
    df_hist.index.name = 'ohdb_id'
    df_hist = df_hist.reset_index()
    df_hist = df_hist.rename(columns={'annsum':'AnnSum','pvol':'Pvol','rx1day':'Rx1day','rx5day':'Rx5day','rx1mon':'Rx1mon','annsum_std':'AnnSum_std','pvol_std':'Pvol_std','rx1day_std':'Rx1day_std','rx5day_std':'Rx5day_std','rx1mon_std':'Rx1mon_std'})
    return (df_hist)

def readGeo():
    df_geo = []
    for varName in ['ai', 'elevation', 'et0', 'GDP', 'arid', 'cold' ,'polar', 'temperate',
                    'tropical', 'cropland', 'forest', 'grassland', 'other', 'urban', 'water', 'wetland', 'ppp', 'slope', 'soilDep']:
        fnames_geo = glob.glob(f'../data/geography/*{varName}_*parquet')[0]
        df_geo0 = pd.read_parquet(fnames_geo).assign(varName=varName)
        df_geo.append(df_geo0)
    df_geo = pd.concat(df_geo)
    df_geo = df_geo.set_index('varName')
    df_geo = df_geo.T
    df_geo.index.name = 'ohdb_id'
    df_geo = df_geo.reset_index()
    return (df_geo)

def readHS():
    df_hs = pd.read_csv('../results/hydrological_signatures_training.csv')
    return df_hs

def merge(df_climate, df_hs, df_geo, df_obs = None):
    df_glh = pd.concat([pd.read_parquet(i) for i in glob.glob('../data/geography/GLH*parquet')])
    df_dam = pd.read_csv('../results/dam_num_training.csv')
    df_riv = pd.concat([pd.read_csv(a) for a in glob.glob('../results/GRIT_reach_length_domain_*.csv')])
    # connect climate, geography, hydrologic signatures
    # delete stations with less than 10 AMS records
    df_obs = df_obs.loc[(df_obs.AMScount>=10)&(df_obs.RL10>0),:].dropna()
    df_train = df_obs[['ohdb_id','AMScount','RL10']].merge(df_hist, on = 'ohdb_id')

    df_train = df_train.merge(df_geo, on = 'ohdb_id')
    df_train = df_train.merge(df_hs, on = 'ohdb_id')
    df_train = df_train.merge(df_glh, on = 'ohdb_id')
    df_train = df_train.merge(df_dam[['ohdb_id','dam_num']], on = 'ohdb_id')
    df_train = df_train.merge(df_riv[['ohdb_id','length_frac']], on = 'ohdb_id', how = 'left')
    df_train['length_frac'] = df_train.length_frac.fillna(0)

    # the shortest distance bewteen gauges and coastlines
    gdf_sta = read_dataframe('../results/OHDB_station_subset.gpkg', read_geometry = False)
    df_train = df_train.merge(gdf_sta[['ohdb_id','dis2sea']], on = 'ohdb_id')

    # new hydrologic signatures: lagT and noResponseRatio (Rain peaks to discharge peaks)
    df_lagT = pd.read_csv('../results/lagT.csv')
    df_train = df_train.merge(df_lagT, on = 'ohdb_id')

    # rename Ks
    df_train = df_train.rename(columns={'logK_Ice_x':'GLHYMPS'})

    # merge lon and lat, darea
    df_train = df_train.merge(gdf_sta[['ohdb_id','ohdb_longitude','ohdb_latitude']],on = 'ohdb_id').rename(columns={'ohdb_longitude':'lon','ohdb_latitude':'lat'})
    gdf_basin = read_dataframe('../data/GRIT/full_catchment/GRIT_full_catchment_all_EPSG8857_500km2_simplify.gpkg', 
                                read_geometry = False)

    df_train = df_train.merge(gdf_basin[['ohdb_id','ohdb_darea','grit_darea']], on = 'ohdb_id')
    df_train['ppp'] = df_train['ppp'].values / df_train.grit_darea.values
    
    df_train['ratio'] = df_train.RL10 / df_train.ohdb_darea
    
    # add climate 
    ds = xr.open_dataset('../data/koppen_5class_0p5.nc')
    df_train['climate'] = df_train.apply(lambda x:float(ds.koppen.sel(lon=x.lon, lat=x.lat, method='nearest').values), axis = 1)
    df_train['climate'] = df_train.climate.map({1:'tropical',2:'dry',3:'temperate',4:'cold',5:'polar'})
    ds.close()

    # add koppen
    ds = xr.open_dataset('../data/geography/koppen_map_1980-2099/1991_2020/koppen_geiger_0p01.nc')
    df_train['koppen'] = np.nan
    da = ds.koppen
    while True:
        df_train['koppen'] = df_train.apply(lambda x:float(da.sel(lon=x.lon, lat=x.lat, method='nearest').values), axis = 1)
        if df_train.koppen.isna().any():
            da = da.interpolate_na(dim = 'lon', method = 'nearest')
        else:
            break
    ds.close()

    return (df_train)

# read parameter name
par_name = pd.read_csv('../data/predictors.csv')

model='ACCESS-ESM1-5'
ens='r1i1p1f1'

# read geography
df_geo = readGeo()

# read hydrologic signatures
df_hs = readHS()

df_hist = readClimate(model, ens, scenario = 'historical', year1 = 1981, year2 = 2014)
fname_obs = '../results/OHDB_500km2_emp_return-level_1981-2014.csv'
df_obs = pd.read_csv(fname_obs)
df_train_st = merge(df_hist, df_hs, df_geo, df_obs = df_obs)
df_train_st = df_train_st.loc[df_train_st.RL10 / df_train_st.ohdb_darea >= 0.01, :]

```

```python
def cleanQ(df):
    # eliminate invalid records
    df1 = df.loc[df.Q.apply(lambda x: not isinstance(x, str)),:]
    df2 = df.loc[df.Q.apply(lambda x: isinstance(x, str)),:]
    try:
        df2 = df2.loc[df2.Q.str.match('\d+'),:]
    except:
        pass
    df = pd.concat([df1, df2])
    df['Q'] = df.Q.astype(np.float32)
    return df

def del_unreliableQ(df):
    '''observations less than 0 were flagged as
        suspected, and (b) observations with more than ten consecutive
        equal values greater than 0 were flagged as suspected'''
    df['date'] = pd.to_datetime(df['date'])
    df = df.sort_values('date').set_index('date')
    index = pd.date_range(df.index[0], df.index[-1], freq = 'D')
    df = df.reindex(index).fillna(0)
    df1 = df.diff()
    df1 = df1.where(df1==0, 1).diff()
    start = np.where(df1.values==-1)[0]
    end = np.where(df1.values==1)[0]
    if len(start) == 0 or len(end) == 0:
        # must no less than zero
        df = df.loc[df.Q>=0,:]
        return (df)
    if start[0] > end[0]:
        start = np.array([0]+start.tolist())
    if start[-1] > end[-1]:
        end = np.array(end.tolist()+[df1.shape[0]+10])
    duration = end - start
    start = start[duration>=10]
    end = end[duration>=10]
    del_idx = np.array([item for a,b in zip(start,end) for item in np.arange(a+1,b+2).tolist()])
    del_idx = del_idx[del_idx<df.shape[0]]
    if len(del_idx) > 0:
        df.drop(df.index[del_idx], inplace = True)
    # must no less than zero
    df = df.loc[df.Q>=0,:]
    return (df)

```

```python
def delete_smaller_sample(df):
    while True:
        intervals = (df.index[1:] - df.index[:-1]).days
        if (intervals > 5).all():
            break
        indices1 = df.index[:-1][intervals<=5]
        indices2 = df.index[1:][intervals<=5]
        for int1,int2 in zip(indices1,indices2):
            if int1 not in df.index:
                continue
            if df.loc[int1] < df.loc[int2]:
                df.drop(int1, inplace=True)
            else:
                df.drop(int2, inplace=True)
    return df

# Function to calculate lag time between rainfall peak and discharge peak
def calc_lagT(pr_dis, p = 95):
    tmp = pr_dis.unstack(level=-1)
    # use scipy to find local discharge peaks
    R95p = tmp.loc[tmp.pr>=0.1,'pr'].quantile(p/100)
    peaks_pr, _ = find_peaks(tmp.pr.values, height=R95p) # find R95p rainfall peaks
    peaks_dis, _ = find_peaks(tmp.dis.values, height=0)
    # remove peaks within five days
    df_peaks_pr = tmp.iloc[peaks_pr,:].pr
    df_peaks_pr = delete_smaller_sample(df_peaks_pr)
    df_peaks_dis = tmp.iloc[peaks_dis,:].dis
    df_peaks_dis = delete_smaller_sample(df_peaks_dis)
    # calculate lag time bewteen rainfall peaks and discharge peaks
    lagT = []
    noResponse = 0
    for index_pr in df_peaks_pr.index:
        index_dis = df_peaks_dis.loc[df_peaks_dis.index>=index_pr]
        if index_dis.shape[0] == 0:
            noResponse += 1
            continue
        index_dis = index_dis.iloc[[0]].index[0]
        if df_peaks_pr.loc[(df_peaks_pr.index>index_pr)&(df_peaks_pr.index<=index_dis)].shape[0] > 0:
            noResponse += 1
            continue
        lagT.append((index_dis - index_pr).days)
    return (pd.Series([np.mean(lagT), noResponse/df_peaks_pr.shape[0]], 
                      index = ['lagT','noResRatio',]))
```
