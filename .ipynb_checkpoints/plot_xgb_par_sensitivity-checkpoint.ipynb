```python
idx = 0
pattern = r'Fitting [0-9] folds for each of [0-9][0-9][0-9] candidates, totalling \d+ fits'
with open(fname,'r') as f:
    for line in f.readlines():
        idx += 1
        if re.match(pattern, line):
            break
```
