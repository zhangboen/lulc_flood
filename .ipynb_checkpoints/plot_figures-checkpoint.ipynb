```python
def add_histogram(ax, vals, extent, vmin, vmax, vind, cmap):
    ax2 = ax.inset_axes(extent)
    cm = plt.cm.get_cmap(cmap)
    bins = [-np.inf] + np.arange(vmin, vmax+vind, vind).tolist() + [np.inf]
    n, bins, patches = ax2.hist(vals, bins=bins)
    bins[0] = vmin - vind
    bins[-1] = vmax + vind
    vcenter=0.3
    norm = mpl.colors.TwoSlopeNorm(vmin=vmin, vmax=vmax, vcenter=vcenter)
    bin_center = 0.5 * (bins[:-1] + bins[1:])
    for c, p in zip(bin_center, patches):
        plt.setp(p, 'facecolor', cm(norm(c)))
    ax2.get_yaxis().set_visible(False)
    ax2.spines['right'].set_visible(False)
    ax2.tick_params(top=False, which='both', labelsize  = 8)
    ax2.xaxis.label.set_visible(False)
    ax2.xaxis.set_minor_locator(ticker.NullLocator())
    ax2.spines['left'].set_visible(False)
    ax2.spines['top'].set_visible(False)
    ax2.spines['right'].set_visible(False)
    ax2.axvline(x=vcenter,color='k')
    ax2.set_xticks([vcenter, vmax])
    
fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (5, 6), subplot_kw={'projection':ccrs.EqualEarth()})
for i,name in enumerate(['Qmin7','Qmax7']):
    if i == 0:
        df = df_Qmin7.copy()
        ax = ax1
        legend = False
    else:
        df = df_Qmax7.copy()
        ax = ax2
        legend = True
    df_sta = df.groupby(['ohdb_id','ohdb_latitude', 'ohdb_longitude','climate_label']).p_apply(
        lambda x: pd.Series(
            [np.corrcoef(x.pred.values,x.Q.values)[0,1]**2], index = ['R2'])
    ).reset_index()
    df_sta['R2_1'] = '< 0.3'
    df_sta.loc[(df_sta.R2>=0.3)&(df_sta.R2<0.5),'R2_1'] = '0.3 - 0.5'
    df_sta.loc[(df_sta.R2>=0.5)&(df_sta.R2<0.7),'R2_1'] = '0.5 - 0.7'
    df_sta.loc[(df_sta.R2>=0.7),'R2_1'] = '>= 0.7'
    title = f'10-fold cross-validation for {name}'
    label = '$\mathregular{R^2}$'
    palette1 = {'< 0.3':'#426dd4','0.3 - 0.5':'#f1eb8b','0.5 - 0.7':'#e47b28','>= 0.7':'#e44f28'}
    sns.scatterplot(df_sta, x = 'ohdb_longitude', y = 'ohdb_latitude', hue = 'R2_1', palette = palette1,
                    hue_order = ['< 0.3', '0.3 - 0.5', '0.5 - 0.7', '>= 0.7'],
                    marker = "$\circ$", ec = "face", alpha = .7, legend = legend, s = 2, 
                    ax = ax, transform = ccrs.PlateCarree())
    if legend:
        sns.move_legend(ax, 'lower left', bbox_to_anchor = (0, 0), title = label, title_fontsize = 9)
    ax.set_global()
    ax.set_ylim([-6525154.6651, 8625154.6651]) 
    ax.set_xlim([-12662826, 15924484]) 
    ax.spines['geo'].set_linewidth(0)
    ax.coastlines(linewidth = .2, color = '#707070')
    ax.add_feature(cf.BORDERS, linewidth = .2, color = '#707070')
    ax.text(0, 1, ['a','b'][i], transform = ax.transAxes, ha = 'center', va = 'top', fontsize = 10, weight = 'bold')
    ax.set_title(title, fontsize = 9)
    # add histograms
    add_histogram(ax, df_sta['R2'].values, [.62, .2, .15, .2], 0, 1, .1, 'RdBu_r')
    
    df_sta = df.groupby(['ohdb_id','ohdb_latitude', 'ohdb_longitude','climate_label']).apply(
            lambda x: pd.Series([
                he.kge(x.pred.values, x.Q.values).squeeze()[0],
                np.corrcoef(x.pred.values, x.Q.values)[0,1],
                np.sum((x.pred.values-x.Q.values)**2) / np.sum(x.Q.values**2) * 100
            ], index = ['KGE','r','nRMSE (%)'])
    )
    df_sta['r'] = df_sta['r'] ** 2
    print(name, df_sta.loc[df_sta.r>=0.3,:].shape[0] / df_sta.shape[0] * 100)
    
ax3 = ax1.inset_axes([1.05, .15, .6, .85])
ax4 = ax2.inset_axes([1.05, .15, .6, .85])
df_Qmin70 = df_Qmin7
plot_scatter(df_Qmin70.Q.values, 
            df_Qmin70['pred'].values, 
            df_Qmin70.climate_label.values, 
            'Observed Qmin7' + '($\mathregular{m^3/s}$)', 
            'Predicted Qmin7' +  '($\mathregular{m^3/s}$)', 
            normColor = 'norm',
            palette = palette,
            fontsize = 7,
            size = 1,
            ax = ax3)
ax3.set_title('Qmin7', fontsize = 9)
df_Qmax70 = df_Qmax7
plot_scatter(df_Qmax70.Q.values, 
            df_Qmax70['pred'].values, 
            df_Qmax70.climate_label.values, 
            'Observed Qmax7' + '($\mathregular{m^3/s}$)', 
            'Predicted Qmax7' +  '($\mathregular{m^3/s}$)', 
            normColor = 'norm',
            palette = palette,
            fontsize = 7,
            legend = False,
            size = 1,
            ax = ax4)
sns.move_legend(ax3, 'upper left', bbox_to_anchor = (0, .99), markerscale = 2, fontsize = 8, handletextpad = .1, borderaxespad = 0)
ax4.set_title('Qmax7', fontsize = 9)

ax3.text(-.2, 1, 'c', weight = 'bold', ha = 'right', va = 'center', fontsize = 10, transform = ax3.transAxes)
ax4.text(-.2, 1, 'd', weight = 'bold', ha = 'right', va = 'center', fontsize = 10, transform = ax4.transAxes)
ax3.set_xlim(-1, 1e5)
ax3.set_ylim(-1, 1e5)
ax4.set_xlim(-1, 1e5)
ax4.set_ylim(-1, 1e5)

# relative bias
ax5 = ax3.inset_axes([1.3, 0, 1, 1])
ax6 = ax4.inset_axes([1.3, 0, 1, 1])
sns.scatterplot(
    df_Qmin7.groupby(['ohdb_id','aridity']).apply(lambda x: pd.Series([(x.pred.mean()-x.Q.mean())/x.Q.mean()*100],index=['bias'])).reset_index(),
    x = 'aridity',
    y = 'bias',
    ax = ax5
)
sns.scatterplot(
    df_Qmax7.groupby(['ohdb_id','aridity']).apply(lambda x: pd.Series([(x.pred.mean()-x.Q.mean())/x.Q.mean()*100],index=['bias'])).reset_index(),
    x = 'aridity',
    y = 'bias',
    ax = ax6
)
for i,ax0 in enumerate([ax5,ax6]):
    ax0.set_xlabel('Aridity index')
    ax0.set_ylabel('Relative bias (%)')
    ax0.set_title(['Qmin7','Qmax7'][i], fontsize = 9)

# fig.tight_layout()
# fig.savefig('../picture/scatter_and_R2_map.png', dpi = 600)
```

### plot trends of urbanization and flood

```python
df_Qmax7 = pd.read_csv('../results/xgb_cv10_Qmax7_seasonal4_multi_MSWX_meteo_raw_result.csv')
df_Qmin7 = pd.read_csv('../results/xgb_cv10_Qmin7_seasonal4_multi_MSWX_meteo_raw_result.csv')

def mk_func(x):
    mk0 = mk.original_test(x)
    return [mk0.slope, mk0.p]

df_Qmax7['climate_label'] = df_Qmax7.climate.map({1:'tropical',2:'dry',3:'temperate',4:'cold',5:'polar'})
df_Qmin7['climate_label'] = df_Qmin7.climate.map({1:'tropical',2:'dry',3:'temperate',4:'cold',5:'polar'})

df_Qmax7_mk = df_Qmax7.groupby(['ohdb_id','ohdb_longitude','ohdb_latitude','climate_label']).p_apply(
    lambda x:pd.Series(
        mk_func(x.ImperviousSurface.values) + mk_func(x.Q.values),
        index = ['slope_urban','p_urban','slope_Q','p_Q']
    )
).reset_index()
df_Qmin7_mk = df_Qmin7.groupby(['ohdb_id','ohdb_longitude','ohdb_latitude','climate_label']).p_apply(
    lambda x:pd.Series(
        mk_func(x.ImperviousSurface.values) + mk_func(x.Q.values),
        index = ['slope_urban','p_urban','slope_Q','p_Q']
    )
).reset_index()
```

### plot model bias with aridity and catchment size

```python
df_Qmin7 = pd.read_csv('../results/xgb_cv10_Qmin7_seasonal4_multi_MSWX_meteo_raw_result.csv')
df_Qmax7 = pd.read_csv('../results/xgb_cv10_Qmax7_seasonal4_multi_MSWX_meteo_raw_result.csv')

df_Qmin7['relbias'] = (df_Qmin7.pred - df_Qmin7.Q) / df_Qmin7.Q * 100
df_Qmax7['relbias'] = (df_Qmax7.pred - df_Qmax7.Q) / df_Qmax7.Q * 100

df_Qmin7['absbias'] = (df_Qmin7.pred - df_Qmin7.Q)
df_Qmax7['absbias'] = (df_Qmax7.pred - df_Qmax7.Q)

df_Qmin7 = df_Qmin7.groupby(['ohdb_id','gritDarea','aridity'])[['relbias','absbias']].mean().reset_index().rename(
    columns={'gritDarea':'Catchment area ($\mathregular{km^2}$)','aridity':'Aridity index',
             'relbias':'Relative bias (%)',
            'absbias':'Absolute bias (mm/day)'})
df_Qmax7 = df_Qmax7.groupby(['ohdb_id','gritDarea','aridity'])[['relbias','absbias']].mean().reset_index().rename(
    columns={'gritDarea':'Catchment area ($\mathregular{km^2}$)','aridity':'Aridity index',
             'relbias':'Relative bias (%)',
             'absbias':'Absolute bias (mm/day)'})

df_Qmin7['Absolute bias (mm/day)'] = df_Qmin7['Absolute bias (mm/day)'] / df_Qmin7['Catchment area ($\mathregular{km^2}$)'] * 86.4
df_Qmax7['Absolute bias (mm/day)'] = df_Qmax7['Absolute bias (mm/day)'] / df_Qmax7['Catchment area ($\mathregular{km^2}$)'] * 86.4
```

```python
fig, axes = plt.subplots(2, 2, figsize = (8, 8))
yname = 'Absolute bias (mm/day)'
sns.scatterplot(df_Qmin7, x = 'Catchment area ($\mathregular{km^2}$)', y = yname, ax = axes[0,0])
sns.scatterplot(df_Qmin7, x = 'Aridity index', y = yname, ax = axes[0,1])
sns.scatterplot(df_Qmax7, x = 'Catchment area ($\mathregular{km^2}$)', y = yname, ax = axes[1,0])
sns.scatterplot(df_Qmax7, x = 'Aridity index', y = yname, ax = axes[1,1])

for i,name in enumerate(['Low flows','High flows']):
    for j in range(2):
        axes[i,j].set_title(name)
        axes[i,j].tick_params(labelsize = 12)
        axes[i,j].set_xlabel(axes[i,j].get_xlabel(), fontsize = 12)
        axes[i,j].set_ylabel(axes[i,j].get_ylabel(), fontsize = 12)
        if i == 0:
#             axes[i,j].set_ylim(-100, 10000)
            axes[i,j].set_ylim(None, 100)
        else:
#             axes[i,j].set_ylim(-100, 6000)
            axes[i,j].set_ylim(None, 100)
fig.tight_layout()
fig.savefig('../picture/absolute_bias_vs_darea_aridity.png', dpi = 600)
```
