```python
import os,glob,sys,re,json,joblib
import numpy as np
import pandas as pd
import xgboost as xgb
import hydroeval as he
from pathlib import Path, PosixPath
from typing import Dict, List, Tuple
from sklearn.model_selection import cross_val_predict, train_test_split, RandomizedSearchCV, KFold
from sklearn.metrics import make_scorer
import pickle
import shap
from datetime import datetime
from tqdm import tqdm
import argparse
from src.utils import check_GPU, LogTrans, InvLogTrans, aleplot_1D_continuous, pdpplot_1D_continuous, undersample
import multiprocessing as mp
import statsmodels.api as sm
from parallel_pandas import ParallelPandas
ParallelPandas.initialize(n_cpu=24, split_factor=24)
from src.plot_utils import *

def load_data(fname, predictors, log = True, m3s = False):
    # read dataset
    df = pd.read_pickle(fname)

    # predictand
    y = df['Q'].astype(np.float32).values
    if log == True and m3s == True:
        raise Exception('log and m3s cannot be both True')
    elif log == True and m3s == False:
        y = LogTrans(y, darea = df['gritDarea'].values, addition = 0.1, log = True)
    elif log == False and m3s == False:
        y = y

    X = df[predictors].astype(np.float32)

    # get feature dtype
    cate_attr = []
    feature_types = X.agg(lambda x: 'q').to_dict()
    for a in ['climate', 'season_id', 'Main_Purpose_id', 'basin_id', 'gauge_id', 'koppen_id', 'year']:
        if a in predictors:
            cate_attr.append(a)
            feature_types[a] = 'c'
            X[a] = X[a].astype(np.int16)
    feature_types = list(feature_types.values())

    # exhaustively search for the optimal hyperparameters
    ml=xgb.XGBRegressor(
        eval_metric='rmsle', 
        tree_method="hist", 
        device='cpu', 
        colsample_bytree = 0.6, 
        learning_rate = 0.1,
        max_depth = 9,
        n_estimators = 1000,
        subsample = 0.9,
        enable_categorical = True,
        feature_types = feature_types,
    )
    
    return df, X, y, ml

def test(fname, predictors):
    # load data
    df, X, y, ml = load_data(fname, predictors)

    cv = KFold(n_splits = 10, shuffle = True, random_state = 1234)
    y_pred = cross_val_predict(ml, X, y, cv = cv)
    y_pred = InvLogTrans(y_pred, darea = df['gritDarea'].values, addition = 0.1, log = True)
    y_pred = np.where(y_pred < 0, 0, y_pred)  

    df['pred'] = y_pred
    df = df[['ohdb_id','ohdb_longitude','ohdb_latitude','season','gritDarea','climate_label','aridity','Q','pred']]
    return df

def effect_raw_func(fname, predictors, group = 'climate_label', log = True, m3s = False, fold = 5):
    # load data
    df, X, y, ml = load_data(fname, predictors)
    
    # split dataframe into 5 splits for each climate region 
    df['split'] = 1
    kf = KFold(n_splits = fold, shuffle = True, random_state = 1234)
    for climate in df.climate_label.unique():
        df0 = df.loc[df.climate_label==climate,:]
        for i,(train_idx, test_idx) in enumerate(kf.split(df0)):
            idx = df0.iloc[test_idx,:].index.values
            df.loc[df.index.isin(idx),'split'] = i+1

    y = df['Q'].astype(np.float32).values
    y = LogTrans(y, darea = df['gritDarea'].values, addition = 0.1, log = True)

    # index of ImperviousSurface for PDP calculation
    index_pdp = predictors.index('ImperviousSurface')

    # cross-validated ALE/PDP
    res_df_all = []
    for split_id in range(1, fold+1):
        X_train = df.loc[df.split!=split_id, predictors]
        idx = X_train.index.values
        y_train = y[idx]

        # refit model
        ml.fit(X_train, y_train)

        train_set_rep = df.loc[(df.split==split_id),:].reset_index()
        res_df = aleplot_1D_continuous(train_set_rep, predictors, ml, 'ImperviousSurface', grid_size = 100, group=group, 
                                        log = log, m3s = m3s, min_interval = 0)
        res_df_all.append(res_df)
    res_df_all = pd.concat(res_df_all)
    return res_df_all
```

### importance of urban areas

```python
meteo_name=['p', 'tmax', 'tmin', 'swd', 'snowmelt', 'smrz']
attr_name=[   'BDTICM', 'elevation', 'slope', 'sedimentary', 'plutonic', 'volcanic', 'metamorphic',
                                    'clay', 'sand', 'silt', 'Porosity_x', 'logK_Ice_x', 
                                    # 'year', 'climate', 'basin_id', 
                                    # 'aridity', 'season_id', 
                                    # 'ohdb_latitude', 'ohdb_longitude', 
                                    'res_darea_normalize', 'Year_ave', 'Main_Purpose_id', 'form_factor', 'LAI',
                        ]
lulc_name = ['ImperviousSurface']

predictors = meteo_name + attr_name + lulc_name
predictors1 = meteo_name + attr_name

out_Qmin7 = test('../data/Qmin7_final_dataset_seasonal4.pkl', predictors)
out_Qmax7 = test('../data/Qmax7_final_dataset_seasonal4.pkl', predictors)

out_Qmin7_no = test('../data/Qmin7_final_dataset_seasonal4.pkl', predictors1)
out_Qmax7_no = test('../data/Qmax7_final_dataset_seasonal4.pkl', predictors1)

Qmin7 = pd.merge(out_Qmin7[['ohdb_id','aridity','climate_label','Q','pred']],
                 out_Qmin7_no[['ohdb_id','aridity','climate_label','Q','pred']],
                 on = ['ohdb_id','aridity','climate_label','Q']
                )
Qmax7 = pd.merge(out_Qmax7[['ohdb_id','aridity','climate_label','Q','pred']],
                 out_Qmax7_no[['ohdb_id','aridity','climate_label','Q','pred']],
                 on = ['ohdb_id','aridity','climate_label','Q']
                )

Qmin7_comp = Qmin7.groupby(['ohdb_id','climate_label','aridity']).p_apply(
    lambda x: pd.Series(
        he.kge(x.pred_x.values, x.Q.values).squeeze().tolist() + [
            np.sqrt(np.mean((x.pred_x.values - x.Q.values)**2)) / (np.max(x.Q.values) - np.min(x.Q.values)) * 100,
            np.sqrt(np.mean((x.pred_x.values - x.Q.values)**2)) / np.mean(x.Q.values) * 100,
        ] + he.kge(x.pred_y.values, x.Q.values).squeeze().tolist() + [
            np.sqrt(np.mean((x.pred_y.values - x.Q.values)**2)) / (np.max(x.Q.values) - np.min(x.Q.values)) * 100,
            np.sqrt(np.mean((x.pred_y.values - x.Q.values)**2)) / np.mean(x.Q.values) * 100,
        ], index = ['KGE','r','alpha','beta','nRMSEminmax', 'nRMSEmean'] + [
            'KGE_no','r_no','alpha_no','beta_no','nRMSEminmax_no', 'nRMSEmean_no'])
).reset_index().melt(id_vars = ['ohdb_id','climate_label','aridity'])
Qmax7_comp = Qmax7.groupby(['ohdb_id','climate_label','aridity']).p_apply(
    lambda x: pd.Series(
        he.kge(x.pred_x.values, x.Q.values).squeeze().tolist() + [
            np.sqrt(np.mean((x.pred_x.values - x.Q.values)**2)) / (np.max(x.Q.values) - np.min(x.Q.values)) * 100,
            np.sqrt(np.mean((x.pred_x.values - x.Q.values)**2)) / np.mean(x.Q.values) * 100,
        ] + he.kge(x.pred_y.values, x.Q.values).squeeze().tolist() + [
            np.sqrt(np.mean((x.pred_y.values - x.Q.values)**2)) / (np.max(x.Q.values) - np.min(x.Q.values)) * 100,
            np.sqrt(np.mean((x.pred_y.values - x.Q.values)**2)) / np.mean(x.Q.values) * 100,
        ], index = ['KGE','r','alpha','beta','nRMSEminmax', 'nRMSEmean'] + [
            'KGE_no','r_no','alpha_no','beta_no','nRMSEminmax_no', 'nRMSEmean_no'])
).reset_index().melt(id_vars = ['ohdb_id','climate_label','aridity'])
```

```python
import pymannkendall as mk
def func(x):
    urban = x.ImperviousSurface.max()
    crop = x.crop.max()
    res = x.res_darea_normalize.max()
    x1 = x.sort_values('year')
    mk0 = mk.original_test(x1.ImperviousSurface.values)
    mk1 = mk.original_test(x1.crop.values)
    out = pd.Series([urban, crop, res, mk0.slope, mk0.p, mk1.slope, mk1.p], index = [
        'urban', 'crop', 'res', 'urban_slope', 'urban_p', 'crop_slope', 'crop_p'
    ])
    return out

def filter_near_natural_catchment(df):
    tmp = df.groupby('ohdb_id').apply(func)
    tmp = tmp.loc[(tmp.urban<5)&(tmp.crop<5)&(tmp.res<5),:].index
    df = df.loc[df.ohdb_id.isin(tmp),:]
    return df
df = pd.read_pickle('../data/Qmax7_final_dataset_annual.pkl')
df = filter_near_natural_catchment(df)
df.ohdb_id.unique().shape
```
