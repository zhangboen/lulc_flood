```python
from src.plot_utils import *
import pickle
import pymannkendall as mk
from scipy import stats
import multiprocessing as mp
from pathlib import Path
import json

dir_Qmax7 = Path('../results/run_Qmax7_onlyUrban_0506_1359_seed824956/')
dir_Qmin7 = Path('../results/run_Qmin7_onlyUrban_0506_1357_seed220973/')

Qmin7Fname = '../data/Qmin7_final_dataset_seasonal4.pkl'
Qmax7Fname = '../data/Qmax7_final_dataset_seasonal4.pkl'

par_map = pd.read_csv('../data/predictors.csv')

with open(dir_Qmax7 / 'cfg.json', 'r') as fp:
    cfg = json.load(fp)
    
mode = cfg['mode']
predictors = cfg['meteo_name'] + cfg['lulc_name'] + cfg['attr_name']
log = cfg['log']
m3s = cfg['m3s']
feature = cfg['feature']
model = cfg['model']
featureName = par_map.loc[par_map.par==feature,'name'].values[0]

with open(dir_Qmin7 / 'cfg.json', 'r') as fp:
    cfg_Qmin7 = json.load(fp)
cfg_Qmax7 = cfg

if mode == 'noLULC':
    predictors1 = [item for item in predictors if item not in ['ImperviousSurface', 'forest', 'crop', 'grass', 'water', 'wetland']]
elif  mode == 'onlyUrban':
    predictors1 = [item for item in predictors if item not in ['forest', 'crop', 'grass', 'water', 'wetland']]

par_map = pd.read_csv('../data/predictors.csv')
```

```python
df0 = pd.read_pickle(cfg_Qmax7['fname'])
shap0 = pickle.load(open(dir_Qmax7 / 'xgb_onlyUrban_shap_values_explain_34581.pkl', 'rb'))
inter0 = pickle.load(open(dir_Qmax7 / 'xgb_onlyUrban_shap_interaction_values_explain_34581.pkl', 'rb'))

idx = predictors.index('ImperviousSurface')

def get_max_inter_idx(x):
    return np.argmax(np.abs(x)[idx,:])

def get_max_inter_value(x):
    idx0 = np.argmax(np.abs(x)[idx,:])
    return x[idx,idx0]

df0['tmp'] = [get_max_inter_idx(inter0[a,:,:]) for a in range(inter0.shape[0])]
df0['max_inter_feature'] = df0['tmp'].apply(lambda x:predictors[x])
df0 = df0.drop(columns=['tmp'])
df0['max_inter_value'] = [get_max_inter_value(inter0[a,:,:]) for a in range(inter0.shape[0])]

# interaction bewteen urban area and soil moisture
idx_smrz = predictors.index('smrz')
df0['inter_urban_smrz'] = inter0[:,idx,idx_smrz]

def get_rank_urban_smrz(x):
    x0 = np.abs(x[idx,:])
    ranks = x0.argsort().argsort()
    rank0 = ranks[idx_smrz]
    return rank0
df0['rank_urban_smrz'] = [get_rank_urban_smrz(inter0[a,:,:]) for a in range(inter0.shape[0])]
```
