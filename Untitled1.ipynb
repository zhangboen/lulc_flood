```python
from src.utils import read_discharge

def func(Ta, P, S_snow0, fdd = 2, Tthr = 1):
    # fdd: melt rate set at 2.0 (mm/d/K) 
    # Tthr: temperature threshold of snowfall
    tmp = fdd * (Ta - Tthr)
    if tmp > S_snow0:
        tmp0 = S_snow0
    else:
        tmp0 = tmp
    if Ta < Tthr: 
        Snowmelt1 = 0
    else:
        Snowmelt1 = tmp0
    tmp1 = S_snow0 - Snowmelt1
    if Ta < Tthr:
        S_snow1 = P + S_snow0
    else:
        if tmp1 > 0:
            S_snow1 = tmp1
        else:
            S_snow1 = 0
    if Ta < Tthr:
        Snowfall1 = P
    else:
        Snowfall1 = 0
    return S_snow1, Snowmelt1, Snowfall1

def main(ohdb_id):
    df_dis = read_discharge(ohdb_id).reset_index()
    df_meteo = pd.read_csv(f'../data_mswx/mswx_each_basin/{ohdb_id}_1981-2023.csv')
    df_meteo['date'] = pd.to_datetime(df_meteo['date'])
    df_meteo = df_meteo.sort_values('date').reset_index(drop=True)
    
    df = pd.merge(df_meteo[['date','pr_mswep','smrz','evap','snowmelt']], df_dis, on = 'date')
    df = df.loc[df.date.dt.year>=1982,:]
    tmp = df.groupby(df.date.dt.year).Q.count()
    # at least 30 years of records during 1982-2020, and each year has at least 300 observations
    tmp = tmp.loc[tmp>=300]
    if tmp.shape[0] < 30:
        return
    df = df.loc[df.date.dt.year.isin(tmp.index),:]
    df = df.drop_duplicates(subset='date').sort_values('date')
    dates = pd.date_range(df.date.values[0], df.date.values[-1], freq = 'D')
    df = df.set_index('date').reindex(dates)
#     # cross-correlation between daily precipitation and discharge for lags of 1-7 days
#     r = [df.pr_mswep.corr(df.Q.shift(a)) for a in range(8)]
#     window = np.argmax(np.array(r))
#     # if the lag is greater than 5 days, it means discharge may not be affected by rainfall, but snow melt
#     if window > 5:
#         return
    window = 7
    # calculate Pr(Q>Q99|P>P99) with the time window
    p99 = df.loc[df.pr_mswep>0.1,'pr_mswep'].quantile(.99)
    q99 = df.Q.quantile(.99)
    df['storm'] = np.where(df.pr_mswep>p99, 1, 0)
    df['flood'] = np.where(df.Q>q99, 1, 0)
    df['smrz7'] = df.rolling(7).smrz.mean()
    df['sm7'] = df.rolling(7).snowmelt.mean()
    df['p7'] = df.rolling(7).pr_mswep.mean()
    df1 = []
    for date in df.loc[df.pr_mswep>p99,:].index:
        df0 = df.loc[(date):(date+pd.Timedelta(days=window)),:]
        storm = df0.loc[date,'pr_mswep']
        storm7 = df0.loc[date,'p7']
        flood = df0.iloc[np.argmax(df0.Q),:]
        date2 = flood.name
        flood = flood.Q
        soil = df0.loc[date,'smrz7']
        snowmelt = df0.loc[date,'sm7']
        evap = df0.evap.mean()
        df1.append(pd.Series([date,storm,storm7,date2,flood,soil,evap,snowmelt], index = [
            'date_storm','storm','storm7d','date_flood','flood','ante_sm7d','evap','ante_snowmelt7d']))
    df1 = pd.concat(df1, axis = 1).T
    df1['date_storm'] = pd.to_datetime(df1['date_storm'])
    df1['date_flood'] = pd.to_datetime(df1['date_flood'])

    # if multiple storm is associated with the same flood, then choose the biggest storm
    df1 = df1.groupby('date_flood').apply(lambda x:x.iloc[np.argmax(x.storm),:]).reset_index(drop=True)

    # connect all storms to calculate the probability
    df1 = df.loc[df.pr_mswep>p99,'pr_mswep'].reset_index().rename(columns={'index':'date_storm'}).merge(
        df1, on = 'date_storm', how = 'left',
    ).drop(columns=['storm']).rename(columns={'pr_mswep':'storm'})

    df1 = df1.rename(columns={'date_storm':'date_p','date_flood':'date_q','storm':'p','flood':'q'})
    df1['q99'] = q99
    df1['p99'] = p99
    df1['window'] = window
    df1['ohdb_id'] = ohdb_id
    df1['sm7d_median'] = df.smrz7.median()
    df1['evap_median'] = df.evap.median()
    df1 = df1.dropna()
    
    # test whether snowmelt-flood ratio
    num0 = df.loc[df.flood==1,:].shape[0]
    if num0 == 0:
        df1['snowmelt_flood_ratio'] = np.nan
    else:
        num = 0
        for date in df.loc[df.flood==1,:].index:
            df0 = df.loc[date,:]
            if df0.p7 < df0.sm7:
                num = num + 1
        ratio = num / num0 * 100
        df1['snowmelt_flood_ratio'] = ratio
    
    cols = ['p','q','ante_sm7d','evap','q99','p99','sm7d_median','evap_median','ante_snowmelt7d']
    df1[cols] = df1[cols].round(3)
    return df1
```

```python
pool = mp.Pool(48)
df_sta = pd.read_csv('../data/basin_attributes.csv')
results = list(tqdm(pool.imap(main, df_sta.ohdb_id.values), total=df_sta.shape[0]))
results = pd.concat(results)
results
    
```
